```python
import os
import shutil

def rename_images(root_folder):
    # Get a list of all image files in the root folder and its subfolders
    image_files = []
    for foldername, _, filenames in os.walk(root_folder):
        for filename in filenames:
            if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
                image_files.append(os.path.join(foldername, filename))

    total_images = len(image_files)

    # Rename the image files sequentially
    for i, old_path in enumerate(image_files):
        folder_path, filename = os.path.split(old_path)
        new_filename = str(i) + '.png'  # Change the extension if needed

        # Create the new path with the same folder structure
        new_path = os.path.join(folder_path, new_filename)

        # Rename the file
        shutil.move(old_path, new_path)

        print(f'Renaming: {old_path} -> {new_path}')

    print('All images renamed successfully.')

# Specify the root folder containing the images
root_folder = 'Image Dataset/Burglary'
rename_images(root_folder)
```
